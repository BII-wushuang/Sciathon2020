Now let's build a quantum circuit to help Ket get to where they need to go! 

(The code is filled in for now, but like the IBM tutorial, this is where the hand-on building of the circuit can come in)

```
def create_circuit():
    qc = qiskit.QuantumCircuit(1)
    qc.h(0)
    return qc

qc = create_circuit()
state = statevec(qc)
plot_state_qsphere(state.data)
```
